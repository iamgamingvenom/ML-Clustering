```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Mall_Customers.csv")
df.head(5)
```

```python
x = df.iloc[:, [3,4]].values
```

```python
df.info()
```

#### Applying elbow method

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.show()
```

#### Applying Kmeans clustering to the dataset

```python
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(x)
```

```python
y_kmeans
```

#### Visualizing the predicted values

```python
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s=100, c='r')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s=100, c='b')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], s=100, c='green')
plt.scatter(x[y_kmeans == 3, 0], x[y_kmeans == 3, 1], s=100, c='cyan')
plt.scatter(x[y_kmeans == 4, 0], x[y_kmeans == 4, 1], s=100, c='blue')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],  s=300, c='yellow', label='centroid')
plt.plot()
```
